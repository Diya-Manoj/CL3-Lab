# Problem Statement:
Implementation of Clonal selection algorithm using Python. 

```python
import random
```

```python
# Fitness Function
def fitness(x):
    return -(x - 5)**2 + 25
```

```python
# Initialize Antibodies
population = [random.randint(0, 10) for _ in range(6)]
```

```python
# Selection (Best Antibodies)
def select(pop):
    return sorted(pop, key=fitness, reverse=True)[:3]
```

```python
# Cloning (Better solutions -> more clones)

def clone(selected):
    clones = []
    for i, val in enumerate(selected):
        num_clones = (len(selected) - i)  # higher fitness → more clones
        clones.extend([val] * num_clones)
    return clones
```

```python
# Hypermutation
def mutate(clones):
    mutated = []
    for c in clones:
        if random.random() < 0.3:
            c += random.randint(-1, 1)
        c = max(0, min(c, 10))  # keep in range
        mutated.append(c)
    return mutated
```

```python
# Main Loop
generations = 5

for gen in range(generations):
    print(f"\nGeneration {gen+1}: {population}")
    
    selected = select(population)
    clones = clone(selected)
    mutated = mutate(clones)
    
    # Combine old + new
    population = sorted(population + mutated, key=fitness, reverse=True)[:6]
```

```python
# Final result
best = max(population, key=fitness)
print("\nBest solution:", best)
print("Best fitness:", fitness(best))
```
