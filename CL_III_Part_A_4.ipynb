# Problem Statement:
Write code to simulate requests coming from clients and distribute them among the servers using the load balancing algorithms. 

### Setup

```python
servers = ["S1", "S2", "S3"]
requests = list(range(1, 11))  # 10 requests
```

### Round Robin Algorithm

```python
print("=== Round Robin ===")

rr_allocation = {}
index = 0

for req in requests:
    server = servers[index % len(servers)]
    rr_allocation[req] = server
    print(f"Request {req} → {server}")
    index += 1
```

### Least Connections Algorithms

```python
print("\n=== Least Connections ===")

# Initialize connection count
connections = {server: 0 for server in servers}
lc_allocation = {}

for req in requests:
    # Find server with least connections
    server = min(connections, key=connections.get)
    
    lc_allocation[req] = server
    connections[server] += 1
    
    print(f"Request {req} → {server} | Load: {connections}")
```

### Request Completion Simulation

```python
import random

connections = {server: 0 for server in servers}

for req in requests:
    server = min(connections, key=connections.get)
    connections[server] += 1
    
    print(f"Request {req} → {server} | Load: {connections}")
    
    # Randomly complete a request
    finished_server = random.choice(servers)
    if connections[finished_server] > 0:
        connections[finished_server] -= 1
print(f"Completed from {finished_server}")
```
