# Problem Statement: 
Implement Union, Intersection, Complement and Difference operations on fuzzy sets. Also create fuzzy relations by Cartesian product of any two fuzzy sets and perform max-min composition on any two fuzzy relations. 

### Define Fuzzy Sets

```python
# Universe
U = ['a', 'b', 'c']

# Fuzzy sets (membership values)
A = {'a': 0.2, 'b': 0.7, 'c': 1.0}
B = {'a': 0.5, 'b': 0.4, 'c': 0.8}
```

### Fuzzy Operations

```python
# Union
union = {x: max(A[x], B[x]) for x in U}

# Intersection
intersection = {x: min(A[x], B[x]) for x in U}

# Complement
complement_A = {x: 1 - A[x] for x in U}
# Use rounding: complement_A = {x: round(1 - A[x], 2) for x in U}

# Difference (A - B)
difference = {x: max(A[x], 1 - B[x]) for x in U}
```

```python
print("Union:", union)
print("Intersection:", intersection)
print("Complement of A:", complement_A)
print("Difference (A - B):", difference)
```

### Cartesian Product (Fuzzy Relation)

```python
relation_R = {}

for u in U:
    for v in U:
        relation_R[(u, v)] = min(A[u], B[v])
```

```python
print("\nFuzzy Relation R (Cartesian Product):")
for key, value in relation_R.items():
    print(f"{key}: {value}")
```

### Max-Min Composition

```python
# Create another relation S 
relation_S = {
    ('a', 'a'): 0.3, ('a', 'b'): 0.6, ('a', 'c'): 0.5,
    ('b', 'a'): 0.7, ('b', 'b'): 0.2, ('b', 'c'): 0.4,
    ('c', 'a'): 0.6, ('c', 'b'): 0.5, ('c', 'c'): 0.3
}
```

```python
# Composition T = R ∘ S
composition = {}

for u in U:
    for w in U:
        values = []
        for v in U:
            val = min(relation_R[(u, v)], relation_S[(v, w)])
            values.append(val)
        composition[(u, w)] = max(values)
```

```python
print("\nMax-Min Composition T:")
for key, value in composition.items():
    print(f"{key}: {value}")
```
