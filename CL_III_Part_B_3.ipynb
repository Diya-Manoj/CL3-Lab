# Problem Statement:
Design and develop a distributed application to find the coolest/hottest year from the available 
weather data. Use weather data from the Internet and process it using MapReduce.

```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("weather.csv")
```

```python
# Convert date column
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
```

```python
# Extract year
df['Year'] = df['Formatted Date'].dt.year
```

```python
# MAP PHASE
mapped = []
for index, row in df.iterrows():
    mapped.append((row['Year'], row['Temperature (C)']))
```

```python
# SHUFFLE & SORT
shuffle = {}
for year, temp in mapped:
    if year not in shuffle:
        shuffle[year] = []
    shuffle[year].append(temp)
```

```python
# REDUCE PHASE
result = {}
for year, temps in shuffle.items():
    result[year] = sum(temps) / len(temps)
```

```python
# FIND HOTTEST & COOLEST
hottest = max(result, key=result.get)
coolest = min(result, key=result.get)
```

```python
print("Yearly Avg Temp:", result)
print("Hottest Year:", hottest)
print("Coolest Year:", coolest)
```
