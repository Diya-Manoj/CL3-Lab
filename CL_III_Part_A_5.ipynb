# Problem Statement:
Optimization of genetic algorithm parameter in hybrid genetic algorithm-neural network 
modelling: Application to spray drying of coconut milk. 

```python
import random
```

```python
def fitness(x):
    return -(x - 5)**2 + 25
```

```python
# Initial Population
population = [random.randint(0, 10) for _ in range(6)]
```

```python
# Selection
def selection(pop):
    pop = sorted(pop, key=fitness, reverse=True)
    return pop[:2]   # best two
```

```python
# Crossover
def crossover(p1, p2):
    return (p1 + p2) // 2
```

```python
# Mutation
def mutation(child):
    if random.random() < 0.2:
        child += random.randint(-1, 1)
    return max(0, min(child, 10))  # restrict between 0–10
```

```python
# GA Loop
generations = 5

for gen in range(generations):
    print(f"\nGeneration {gen+1}: {population}")
    
    p1, p2 = selection(population)
    
    new_population = [p1, p2]
    
    while len(new_population) < len(population):
        child = crossover(p1, p2)
        child = mutation(child)
        new_population.append(child)
    
    population = new_population
```

```python
# Final result
best = max(population, key=fitness)
print("\nBest solution:", best)
print("Best fitness:", fitness(best))
```
